Connet duckdb in python

```python
import duckdb 

with duckdb.connect("database/jokes.duckdb") as conn:
    conn.sql("""--sql
    CREATE OR REPLACE TABLE jokes (
        id INT,
        joke STRING,
        );
    """)

    conn.sql("""--sql
    INSERT INTO jokes VALUES
        (1, 'Why do programmer hate sunlight? cus it attraches bugs'),
        (2, 'why do javaprogrammers need glasses. cus they have cannot c#');
    """)

    desc = conn.sql("DESC;").df()
```
